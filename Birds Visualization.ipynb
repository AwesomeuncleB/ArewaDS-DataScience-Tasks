```python
# Import Pandas
import pandas as pd

birds_df = pd.read_csv('birds.csv')

birds_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set(style="whitegrid")

# Count the number of bird species within each conservation status
conservation_status_counts = birds_df['ConservationStatus'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=conservation_status_counts.index, y=conservation_status_counts.values, palette="coolwarm")
plt.title('Number of Bird Species by Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Number of Species')
plt.xticks(rotation=45)
plt.show()
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

# Assuming birds_df is your DataFrame and average_max_body_mass has been calculated as shown before
# Calculate the average maximum body mass by category
average_max_body_mass = birds_df.groupby('Category')['MaxBodyMass'].mean()

# Sort the values and select the top ten
top_ten_average_max_body_mass = average_max_body_mass.sort_values(ascending=False).head(10)

# Set the figure size for better visibility
plt.figure(figsize=(12, 8))

# Use seaborn's barplot for a more aesthetically pleasing look. This also allows for easy customization.
sns.barplot(x=top_ten_average_max_body_mass.values, y=top_ten_average_max_body_mass.index, palette="viridis")

# Adding titles and labels for clarity
plt.title('Top 10 Bird Categories by Average Maximum Body Mass')
plt.xlabel('Average Maximum Body Mass (g)')
plt.ylabel('Category')

# Inverting the y-axis to have the category with the highest average at the top
plt.gca().invert_yaxis()

plt.show()



```

```python
# 2. Distribution of Maximum Wingspan
# Plotting the histogram of maximum wingspan
plt.figure(figsize=(10, 6))
sns.histplot(birds_df['MaxWingspan'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Maximum Wingspan')
plt.xlabel('Maximum Wingspan (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a scatter plot comparing MaxBodyMass and MaxWingspan, colored by ConservationStatus
plt.figure(figsize=(12, 8))

# Using seaborn to create the scatter plot with conservation status as hue
sns.scatterplot(data=birds_df, x='MaxBodyMass', y='MaxWingspan', hue='ConservationStatus', style='ConservationStatus', palette='viridis', s=100)

plt.title('Body Mass vs. Wingspan by Conservation Status')
plt.xlabel('Maximum Body Mass (g)')
plt.ylabel('Maximum Wingspan (cm)')
plt.legend(title='Conservation Status', bbox_to_anchor=(1.05, 1), loc='upper left')

# Log scale for better visualization
plt.xscale('log')

plt.tight_layout()
plt.show()
```
